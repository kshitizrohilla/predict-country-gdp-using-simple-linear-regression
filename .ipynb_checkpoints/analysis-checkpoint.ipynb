# Predicting country's gdp using simple linear regression
Given Canada's adjusted net per capita income determine the per capita income of a year say 2020 or 2022 (say).

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import linear_model

import warnings
warnings.filterwarnings('ignore', message='X does not have valid feature names')
```

```python
df = pd.read_csv("data/canada-per-capita-income.csv")
df.head()
```

```python
print(df.dtypes)
```

Creaing a scatter plot of per-capita-income vs year

```python
plt.scatter(df["year"], df["per_capita_income"], color="dodgerblue")

plt.xlabel("Year")
plt.ylabel("Per-Capita-Income (in $)")

plt.show()
```

Creaing a line graph of per-capita-income vs year

```python
plt.plot(df["year"], df["per_capita_income"], color="dodgerblue")

plt.xlabel("Year")
plt.ylabel("Per-Capita-Income (in $)")

plt.show()
```

```python
reg_obj = linear_model.LinearRegression()
reg_obj.fit(df[["year"]], df["per_capita_income"])
```

```python
df2 = pd.DataFrame({"year": df["year"]})
df2.head()
```

```python
predicted_pci = reg_obj.predict(df2)
print(predicted_pci)
```

```python
df["predicted_per_capita_income"] = predicted_pci
df.head()
```

```python
m = reg_obj.coef_
print(m) # outputs the value of slope(m)
```

```python
c = reg_obj.intercept_
print(c) # outputs the intercept value(c)
```

```python
plt.scatter(df["year"], df["predicted_per_capita_income"], color="red")
plt.scatter(df["year"], df["per_capita_income"], color="dodgerblue")

plt.xlabel("Year")
plt.ylabel("Per-Capita-Income (in $)")

plt.show()
```

```python
plt.plot(df["year"], df["predicted_per_capita_income"], color="red")
plt.scatter(df["year"], df["per_capita_income"], color="dodgerblue")

plt.xlabel("Year")
plt.ylabel("Per-Capita-Income (in $)")

plt.show()
```

```python
x = df["year"].to_numpy()
print(x)
```

```python
y = m * x + c
using_formula_arr = np.empty(len(x))
for i in range(len(x)):
    using_formula_arr[i] = m * x[i] + c
    
print(using_formula_arr, end=", ")
```

```python
np.array_equal(predicted_pci, using_formula_arr)
```
